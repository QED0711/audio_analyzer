```python
%load_ext autoreload
%autoreload 1

from python.AudioAnalyzer import *
%aimport python.AudioAnalyzer
```

```python
original = AudioAnalyzer("audio/yesterday-original.wav", input_sr=44100, fft_size=44100)
remaster_09 = AudioAnalyzer("audio/yesterday-2009.wav", input_sr=44100, fft_size=44100)
remaster_15 = AudioAnalyzer("audio/yesterday-2015.wav", input_sr=44100, fft_size=44100)
```

```python
original.plot_spectrum(min_freq=20, max_freq=1000, title="1965 Original")
remaster_09.plot_spectrum(min_freq=20, max_freq=1000, title="2009 Remaster")
remaster_15.plot_spectrum(min_freq=20, max_freq=1000, title="2015 Remaster")
```

```python
orig_to_09 = SpectrumCompare(original, remaster_09)
orig_to_15 = SpectrumCompare(original, remaster_15)
remasters = SpectrumCompare(remaster_09, remaster_15)
```

```python
# orig_to_09.plot_spectrum_group(frange=(20,1000), title="Spectrograms - original vs 2009 remaster", legend=("Original", "Remaster"))
orig_to_09.plot_spectrum_group(frange=(20,1000), 
                               ratio=True, 
                               threshold=True,
                               title="Spectrograms - original vs 2009 remaster (with threshold)", 
                               legend=("Original", "Remaster", "Amplitude Diff.", "Threshold")
                              )
orig_to_09.plot_spectrum_heatmap(
    frange=(20,1000), 
    plot_spec1=False,
    title="Original vs 2009 remaster"
)
orig_to_09.plot_spectrum_heatmap(
    frange=(20,500), 
    plot_spec1=False, 
    title="Original vs 2009 remaster"
)
```

```python
orig_to_15 = SpectrumCompare(original, remaster_15)

orig_to_15.plot_spectrum_group(frange=(20,1000), 
                               ratio=True, 
                               threshold=True,
                               title="Original vs 2015 Remaster",
                               legend=("Original", "Remaster", "Amplitude Diff.", "Threshold")
                              )
orig_to_15.plot_spectrum_heatmap(frange=(20,500), 
                                 plot_spec1=False, 
                                 cmap='magma',
                                 title="Original vs 2015 Remaster Heatmap (20-500 Hz)"
                                )
```

```python
remasters = SpectrumCompare(remaster_09, remaster_15)

remasters.plot_spectrum_group(frange=(20,1000), 
                              ratio=True, 
                              threshold=True,
                              title="2009 vs 2015 Remaster",
                              legend=("2009", "2015", "Amplitude Diff.", "Threshold")
                             )
remasters.plot_spectrum_heatmap(frange=(20,200), 
                                plot_spec1=True, 
                                cmap='viridis',
                                title="2009 vs 2015 Heatmap (20-200 Hz)"
                               )
```
